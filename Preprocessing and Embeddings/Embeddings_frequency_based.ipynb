# **Bag of Words (BoW)**

---



BoW technique which creates a vector of counts by counting the occurence of each token in a corpus, it disregards the position of the words. It is helpful in tasks like classification.

---



```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

bow = cv.fit_transform(new_df)
```

```python
cv.vocabulary_
```

```python
bow.toarray()
```

# **N-Grams**

---

N-Grams are contiguous sequence of n items froma given text. A simple tokenization can be regarded as unigram, while grouping of two words is called bigram and so on. This technique is used for tasks like text prediction and auto complete feature.

---



```python
cv2 = CountVectorizer(ngram_range=(3,3))

bow2 = cv2.fit_transform(new_df)

cv2.vocabulary_
```

# **Term Frequency - Inverse Document Frequency (TF-IDF)**

---
TF-IDF is a statistical method to determine importance of a word based on its appearance in a document. It calculates term frequency (TF) first, which shows how often a word appears in a document. Then it calculates inverse document frequency (IDF) which is a measure of rarity of that word in a series of documents. Multiplying these two values i.e. TF and IDF gives out a measurement for importance of words. A word occurring multiple times in a particular document and rarely in other documents will be given a high score.
TF-IDF finds its application in tasks like information retrieval and search engines.

---



```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfid = TfidfVectorizer()

arr = tfid.fit_transform(new_df)

arr.toarray()
```
